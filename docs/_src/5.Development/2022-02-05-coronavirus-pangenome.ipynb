````bash
DIR="$HOME/workspace/2022-01-31-coronavirus-metapangenome-graph"
mkdir -p $DIR
cd $DIR
````

````bash
DB_PATH="$DIR/coronavirus_db"
````

````bash
ls -1 $DIR/ncbi-files/*.fna.gz > $DIR/genomes.txt
````

````bash
export pantools="java -Xms1g -Xmx8g -jar $HOME/git/pantools/dist/pantools.jar"
````

````bash
sudo conda run -n pantools $pantools build_pangenome -dp $DB_PATH -gf $DIR/genomes.txt
````

````bash
gzip -d ncbi-files/*.gff.gz
````

````bash
ls -1 ncbi-files/*.gff | awk -F "i=1" '{print ++i, $0}' > annotation_locations.txt
````

````bash
sudo conda run -n pantools $pantools add_annotations -dp $DB_PATH -af annotation_locations.txt -ca
````

````bash
sudo conda run -n pantools $pantools group -h
````

Functional classification

Similar to gene and k-mer classification, this function identifies core, accessory, unique functional annotations in the pangenome. Only the following functions are considered for this analysis: biosynthetic gene clusters from antiSMASH, GO, PFAM, InterPro, TIGRFAM.
